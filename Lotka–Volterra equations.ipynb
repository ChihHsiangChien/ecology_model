 # Lotka–Volterra equations
 ## Predator Prey Model
 ### R<sub>n</sub> = R<sub>n-1</sub>  + *a*R<sub>n-1</sub> - *b*R<sub>n-1</sub>F<sub>n-1</sub> 
 ### F<sub>n</sub> = F<sub>n-1</sub>   - *c*F<sub>n-1</sub> + *d*R<sub>n-1</sub>F<sub>n-1</sub> 
 
 
 

```python
import numpy as np
import matplotlib.pyplot as plt 

n = 5000    #幾代

rabbits = np.zeros(n)
foxes = np.zeros(n)

#initial
rabbits[0] = 3000
foxes[0] = 4000

a = 0.005     #rabbits birth
b = 5e-6      #捕食機率
c = 0.006     #foxes death
d = 5e-6      #捕食機率2

#Recurrances
for i in range(1,n):
    rabbits[i] = rabbits[i-1] + a * rabbits[i-1] - b * rabbits[i-1] * foxes[i-1]
    foxes[i]   = foxes[i-1]   - c * foxes[i-1]   + d * rabbits[i-1] * foxes[i-1]

plt.plot(rabbits)
plt.plot(foxes)
#plt.ylim(0,10000)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def f(Y, t):
    y1, y2 = Y
    return [y2, -np.sin(y1)]

y1 = np.linspace(-2.0, 8.0, 20)
y2 = np.linspace(-2.0, 2.0, 20)

Y1, Y2 = np.meshgrid(y1, y2)

t = 0

u, v = np.zeros(Y1.shape), np.zeros(Y2.shape)

NI, NJ = Y1.shape

for i in range(NI):
    for j in range(NJ):
        x = Y1[i, j]
        y = Y2[i, j]
        yprime = f([x, y], t)
        u[i,j] = yprime[0]
        v[i,j] = yprime[1]
     
plt.quiver(Y1, Y2, u, v, color='r')

plt.xlabel('$y_1$')
plt.ylabel('$y_2$')
plt.xlim([-2, 8])
plt.ylim([-4, 4])
plt.show()
```
